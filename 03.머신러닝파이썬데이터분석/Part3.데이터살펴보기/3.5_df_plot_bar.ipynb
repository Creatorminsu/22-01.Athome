```python
import pandas as pd
```

```python
df = pd.read_excel('./남북한발전전력량.xlsx', engine='openpyxl') # 데이터 프레임 변환
```

```python
# 남한, 북한 발전량 합계 데이터만 추출
df_ns = df.iloc[[0,5], 3:]
df_ns.index = ['South','North']
df_ns.columns = df_ns.columns.map(int)
```

```python
# 행, 열 전치하여 막대 그래프 그리기
tdf_ns = df_ns.T
print(tdf_ns.head())
print('\n')
tdf_ns.plot(kind='bar')
```
