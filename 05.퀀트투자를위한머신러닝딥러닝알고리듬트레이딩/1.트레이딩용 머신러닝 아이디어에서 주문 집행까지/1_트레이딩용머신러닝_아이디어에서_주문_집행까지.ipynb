퀀트투자를위한 머신러닝 딥러닝 알고리듬 트레이딩
1. 머신러닝 능력과 대체데이터에 대한 투자로 추진
- 우호적인 가격결정을 목표로 하는 거래 실행 알고리듬
- 작은 가격변동(예를 들면 차익거래)으로 부터 이익을 얻을 목적으로 하는 단기거래
- 다른 시장 참여자의 행동을 예측하는 것을 목표로 하는 행태거래 전략
- 절대 및 상대 가격을 기반으로 하는 거래 전략과 수익률 예측
​
2. 머신러닝 사용 사례
- 패턴을 확인하고 특성(feature)를 추출해 통찰력을 생성하는 데이터 마이닝
- 리스크 팩터나 알파와 거래 아이디어를 창출하기 위한 지도 학습
- 개별 신호를 전략으로 통합
- 알고리듬으로 학습된 리스크 프로파일에 따른 자산 배분
- 합성 데이터의 사용이 포함된 전략의 테스트와 평가
- 강화학습을 사용한 자동화된 대화형 전략 개선
​
3. 특성 추출과 통찰력을 위한 데이터 마이닝
- 정보이론(information theory)은 후보 특성의 신호 내용 평가를 도우며, 머신러닝 모델에 대한
가장 중요한 입력 변수를 추출하는데 유용하다. 4장에서 자산 수익률 예측을 위한 지도학습 알고리듬에 대해
개별 특성의 잠재적인 가치를 평가하고자 상호 정보량(mutual information)을 사용한다.
데 프라도(De Prado 2018)의 18장에서는 대체 거래 전략을 결정하기 위한 기준으로 가격 시계열의 정보내용을 추정한다.
- 비지도 학습은 데이터 구조를 식별해 통찰력을 얻거나 다운 스트림 작업을 해결하는데 도움이 되는 광범위한 방법을 제공한다. 클러스터링, 차원축소, 베이지안 확률모델을 적용
딥러닝을 사용해 자산 특성에 따라 조건화된 비선형 리스크 팩터를 추출하고 켈리 등(2020)을 기반으로 주식 수익률을 예측한다.
-모델 투명성 개별 변수의 예측력에 대한 통찰력을 얻고 SHAP(SHapley Addictive exPlanations)라는 새로운 게임 이론적 접근 방식을 도입해 모델 고유의 방법을 강조한다. 
- 다운스트림 모델(downstream model) 개별 자산의 전망, 자본 시장 기대, 증권 간의 상관관계에 대한 예측
- 임의 일임 거래에 영향을 미칠 수 있다.
